```python
import numpy as np 
import matplotlib.pyplot as plt 

def sigmoid(x):
    return 1/(1+np.exp(-x))


sigmoid_values=[]

for val in np.arange(-1000.0,1000.0,0.5):
    sigmoid_values.append(sigmoid(val))
    
plt.plot(np.arange(-1000,1000,0.5),sigmoid_values)
plt.show()
```

```python
sigmoid_derivatives=[]

for val in np.arange(-1000.0,1000.0,0.05):
    sigmoid_derivatives.append(sigmoid(val)*(1-sigmoid(val)))
    
    
plt.plot(np.arange(-1000.0,1000.0,0.05),sigmoid_derivatives)

plt.ylabel('Sigmoid Derivative Values')
plt.xlabel('Values')
```
